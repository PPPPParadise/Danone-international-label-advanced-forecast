```python
import pandas as pd
import numpy as np
import datetime
from plotly.graph_objs import Scatter,Layout
import plotly
import plotly.offline as py
import plotly.graph_objs as go

#setting offilne
plotly.offline.init_notebook_mode(connected=True)
```

```python
Off_take_Forecast = pd.read_excel('U1_Offtake_Forecast.xlsx',sheet_name= 'TTL')
```

```python
Off_take_Forecast.replace(np.nan,0,inplace = True)
```

```python
def get_month_range(start_day,end_day):
    months = (end_day.year - start_day.year)*12 + end_day.month - start_day.month
    month_range = ['%s-%s'%(start_day.year + mon//12,mon%12+1) 
                    for mon in range(start_day.month-1,start_day.month + months)]
    return month_range
```

```python
x0 = get_month_range(datetime.date(2017,1,1),datetime.date(2021,12,31))
```

```python
y1 = Off_take_Forecast.loc[1,:][1:].to_numpy().astype(np.int)
```

```python
# Create traces
fig = go.Figure()

for i in range(Off_take_Forecast.index.min(),4):
    y0 = Off_take_Forecast.loc[i,:][1:].to_numpy().astype(np.int)
    fig.add_trace(go.Scatter(x=x0, y=y0,
                    mode='lines+markers', name=Off_take_Forecast['in tin'][i]))

fig.update_layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=False,
        showticklabels=False,
    ),
    plot_bgcolor='white'
)

fig.show()
```
